```python
import pandas as pd
covid_df = pd.read_csv('C:\\Users\\admin2\\Documents\\Python\\DataScience_New\\DataFrame\\covid\\covid.csv')
covid_df.head()
```

```python
grouped_cases = covid_df.groupby('date')['daily_confirmed'].sum()
grouped_cases.plot(
    kind='line',
    figsize=(12, 4),
    title='Ежедневная заболеваемость по всем странам',
    grid = True,
    lw=3
)
```

```python
#группируем по странам заболевших, берем последний элемент
grouped_cases = covid_df.groupby('country')['confirmed'].last()
#выборем 10 наибольших значений
grouped_cases = grouped_cases.nlargest(10)
grouped_cases.plot(
    kind='bar',
    figsize=(12, 4),
    colormap='plasma',
    grid = True
)
```

```python
grouped_cases = covid_df.groupby('country')['total_vaccinations'].last()
#выборем 10 наибольших значений
grouped_cases = grouped_cases.nsmallest(10)
grouped_cases.plot(
    kind='bar',
    figsize=(20, 4),
    colormap='plasma',
    grid = True
)
```
