```python
import pandas as pd
import plotly.express as px
```

```python
churn_data = pd.read_csv('C:\\Users\\admin2\\Documents\\Python\\DataScience_New\\DataFrame\\covid\\churn.csv')
churn_data.head()
```

```python
churn_data = churn_data.drop(columns=['RowNumber'])
churn_data.head()
```

```python
#Задание 9.1
fig = px.bar(
    churn_data, #датафрейм
    x="Exited", #ось x
    text = 'Exited', #текст на столбцах
    orientation='v', #ориентация графика
    height=500, #высота
    width=1000, #ширина
    title='Соотношение лояльности клиентов' #заголовок
)

#отображаем его
fig.show()
```

```python
fig = px.pie(churn_data, values='Exited', names='Exited', title='Соотношение лояльности клиентов')
fig.update_layout(title_x=0.48)

fig.show()
```

9.1. Каково соотношение ушедших и лояльных клиентов? Покажите это на графике и дайте комментарий по соотношению.

```python
#Задание 9.1
Exited_pie = churn_data['Exited'].value_counts()
fig = plt.figure(figsize=(5, 5))
axes = fig.add_axes([0, 0, 1, 1])
axes.pie(
    Exited_pie,
    labels=Exited_pie.index,
    autopct='%.1f%%',
    explode = [0, 0]
)
```

9.2. Постройте график, показывающий распределение баланса пользователей, у которых на счету больше 2 500 долларов. Опишите распределение и сделайте выводы.

```python
fig2 = px.line(
    churn_data, #датафрейм
    y="Balance", #ось x
    text = 'Balance', #текст на столбцах
    orientation='h', #ориентация графика
    height=500, #высота
    width=1000, #ширина
    title='Соотношение лояльности клиентов' #заголовок
)

#отображаем его
fig2.show()
```

```python
line_data = churn_data.groupby('Balance', as_index=False).sum()
fig = px.line(
    data_frame=line_data, #DataFrame
    #x='Date', #ось абсцисс
    y=['Balance'], #ось ординат
    height=500, #высота
    width=1000, #ширина
    title='Confirmed, Recovered, Deaths, Active cases over Time' #заголовок
)
fig.show()
```
