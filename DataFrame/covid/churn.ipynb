```python
import pandas as pd
import plotly.express as px
```

```python
churn_data = pd.read_csv('C:\\Users\\admin2\\Documents\\Python\\DataScience_New\\DataFrame\\covid\\churn.csv')
churn_data.head()
```

```python
churn_data = churn_data.drop(columns=['RowNumber'])
churn_data.head()
```

```python
# 9.1
fig = px.bar(
    churn_data, #
    x="Exited", # x
    text = 'Exited', #  
    orientation='v', # 
    height=500, #
    width=1000, #
    title='  ' #
)

# 
fig.show()
```

```python
fig = px.pie(churn_data, values='Exited', names='Exited', title='  ')
fig.update_layout(title_x=0.48)

fig.show()
```

```python
# 9.1
Exited_pie = churn_data['Exited'].value_counts()
fig = plt.figure(figsize=(5, 5))
axes = fig.add_axes([0, 0, 1, 1])
axes.pie(
    Exited_pie,
    labels=Exited_pie.index,
    autopct='%.1f%%',
    explode = [0, 0]
)
```
