```python
import pandas as pd
orders_df = pd.read_csv('C:\\Users\\admin2\\Documents\\Python\\DataScience_New\\DataFrame\\orders.csv' , sep=';')
product_df = pd.read_csv('C:\\Users\\admin2\\Documents\\Python\\DataScience_New\\DataFrame\\orders.csv', sep=';')
orders_df.info()
```

```python
#слияние двух файлов
orders_products = orders_df.merge(
    product_df,
    on='ID товара',
    how='left'
)
display(orders_products.head())
```

```python
mask = (orders_products['Order ID'] > 0)
display(orders_products = orders_products[mask])
```

```python
#orders_products.tail(1)['Order ID_x']
#orders_products[orders_products['Отменен_x'] == 'Да']
orders_products['Profit'] = orders_products['Price'] * orders_products['Количество'] 
orders_products[orders_products['Оплачен'] == 'Да'].groupby('ID Покупателя')['Profit'].sum().sort_values(ascending=False)
```
